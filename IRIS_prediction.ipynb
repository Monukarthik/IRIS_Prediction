```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=pd.read_csv('Iris.csv')
a.sample(10)
```

```python
b=a.drop(["Id"],axis=1)
b.boxplot()
```

**Removing Outlers**

```python
q1=b['SepalWidthCm'].quantile(0.25)
q3=b['SepalWidthCm'].quantile(0.75)
iqr=q3-q1
l_bound=q1-1.5*iqr
u_bound=q3+1.5*iqr
filtered=b[(b['SepalWidthCm']>=l_bound)&(b['SepalWidthCm']<=u_bound)]
filtered
```

```python
filtered.shape
```

```python
filtered.value_counts(["Species"])
```

```python
filtered.head()
```

**Model Training**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from  sklearn.metrics import accuracy_score,classification_report
```

```python
x=filtered.drop('Species',axis=1)
y=filtered['Species']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
```

```python
accuracy_score(y_test,ypred)
```

```python
print(classification_report(y_test,ypred))
```
